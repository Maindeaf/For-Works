```python
#建立所需方法
def convert(seconds):
    hour = seconds // 3600
    seconds %= 3600
    minutes = seconds // 60
    seconds %= 60
    return "%04d:%02d:%02d" % (hour, minutes, seconds)
def splittime(time):
    timesplitlist = []
    timesplitlist = re.split(':', time)
    seconds = timedelta(hours = int(timesplitlist[0]),
              minutes = int(timesplitlist[1]),
              seconds = int(timesplitlist[2])).total_seconds()
    return seconds
```
